```python
import markov
from konlpy.tag import Okt
import json
```

```python
ballad_file = "../../Note/Markov/ballad.txt"
dict_file = "../../Note/Markov/markov_ballad.json"
```

```python
file = open("../../Data/ballad_lyrics_happy.csv")
text = str(file.readlines())
text
```

```python
text = text.split(",")[1:]
```

```python
text
```

```python
total = ""
for i in range(len(text)):
     total += " " + text[i]
```

```python
total
```

```python
# 텍스트를 한 줄씩 처리하기
twitter = Okt()
lines = total.split("\n")
for line in lines:
    malist = twitter.pos(line,norm=True,stem=True)
    words = []
    for word in malist:
        if not word[1] in ['Alpha','Punctuation']: # .(마침표)제외한 ? "" ! 를 사용하지 않음
            words.append(word[0])
        if word[0] == ".":
            words.append(word[0])
```

```python
# 딕셔너리 생성
dic = markov.make_dic(words)
json.dump(dic,open(dict_file,"w",encoding="utf-8"))
```

```python
dic = json.load(open(dict_file,"r"))
```
