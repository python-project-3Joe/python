```python
import pandas as pd
from konlpy.tag import Okt
from gensim.models import word2vec
```

```python
indif = pd.read_csv("../../Data/ballad_indifference20.csv")
```

```python
indif.loc[0,'가사']
```

```python
# 텍스를 한 줄씩 처리하기
twitter = Okt()
results = []
lines = indif.loc[0,'가사']
for line in lines:
    # 형태소 분석하기
    malist = twitter.pos(line,norm=True,stem=True) # 단어의 기본형 사용
    r = []
    for word in malist:
        # 어미, 조사, 구두점 등은 대상에서 제외
        if not word[1] in ["Josa","Eomi","Punctuation"]:
            if word[0] != "n":
                r.append(word[0]) # word[0] : data
    rl = (" ".join(r).strip())
    results.append(rl)
    print(rl)
```
