```python
import pandas as pd
import nltk # national language tool kit
from konlpy.corpus import kolaw
from konlpy.tag import Okt
import matplotlib.pyplot as plt
from wordcloud import WordCloud

t = Okt()
```

```python
ballad_regret20 = pd.read_csv("../../Data/ballad_regret20.csv")
ballad_regret20.head()
```

---
### 후회 wordcloud 생성(명사)

```python
regret20 = ""
```

```python
for i in range(0, 20):
    regret20 += (" " + ballad_regret20['가사'][i])
print(regret20)
```

```python
# 명사 분리
token_regret = t.nouns(regret20)
```

```python
# 후회테마 20곡의 상위 50개 단어 확인
# 이걸 보면서 불용어 선택하기
token_regret50 = nltk.Text(token_regret)
token_regret50.vocab().most_common(50)
```

```python
# 불용어 추가
# 실행해보면서 추가하면서 바꿔야함.
stop_words20 = ['내','말','그','수','네','날','널','그대','니','햇살','해','이','걸','것','볼','거','난','꼭','게','길','길','알','안','눈','줄','척','속','때','손','늘','꿈','전','만큼','뒤','젠']
token_regret50 = [each_word for each_word in token_regret if each_word not in stop_words20]
```

```python
# 후회테마 20곡의 상위 50개 단어 확인
# 불용어 처리후
token_regret50 = nltk.Text(token_regret50)
token_regret50.vocab().most_common(50)
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# 상위 50개의 빈도수 차트 만들기
plt.figure(figsize=(12,6))
token_regret50.plot(50)
plt.show()
```

```python
# 상위 50개만 wordcloud로 그리기
# 구한데이터로 그리기
data50 = token_regret50.vocab().most_common(50)

# 딕셔너리로 만들기
tmp_data50 = dict(data50)

wordcloud50 = WordCloud(
    background_color='White',
    relative_scaling=0.2,
    font_path="AppleGothic" # 윈도우 => 'c:/Windows/Fonts/malgun.ttf'
).generate_from_frequencies(tmp_data50)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(wordcloud50, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## 후회 워드클라우드 - 동사

```python
test = regret20
```

```python
verb_list = []
for i in range(len(t.pos(test))):
    if t.pos(test)[i][1] == 'Verb':
        verb_list.append(t.pos(test)[i][0])
```

```python
verb_list
```

```python
# 불용어 추가
# 실행해보면서 추가하면서 바꿔야함.
stop_words2 = ["할","참","한","챌","는제","참","줘","걸까","댄","따라","된","잘","된","는","올","되",\
    "기를","까","큰","갈","건지","하는","이길","딴","서","써","내게","나를","나도",'해서','댈','봐','하며','해','했죠','해요','했어','본','밝아서','줬던','돼']
token_ko2 = [each_word for each_word in verb_list if each_word not in stop_words2]
print(token_ko2)
```

```python
# 상위 50개 단어 확인
# 이걸보면서 불용어 지우는게 편해 전체는 힘들어
ko2 = nltk.Text(token_ko2)
ko2.vocab().most_common(50)
```

```python
# 상위 50개의 빈도수 차트 만들기
plt.figure(figsize=(12,6))
ko2.plot(50)
plt.show()
```

```python
# 상위 50개만 wordcloud로 그리기
# 구한데이터로 그리기
data2 = ko2.vocab().most_common(100)

# 딕셔너리로 만들기
tmp_data2 = dict(data2)

wordcloud2 = WordCloud(
    background_color='White',
    relative_scaling=0.8,
    font_path="AppleGothic" # 윈도우 => 'c:/Windows/Fonts/malgun.ttf'
).generate_from_frequencies(tmp_data2)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis('off')
plt.show()
```
