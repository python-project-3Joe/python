## ballad_sad에서 20개 데이터 추출하기

```python
import pandas as pd
import numpy as np
import random
```

```python
sad = pd.read_csv("../../Data/ballad_sad.csv")
len(sad)
```

```python
index_list = range(0,30)
```

```python
index = random.sample(index_list,20)
index
```

```python
sad20 = sad.iloc[index,:]
```

```python
sad20.reset_index(drop=True,inplace=True)
```

```python
sad20.to_csv("../../Data/ballad_sad20.csv",index=False)
```

```python
sad20 = pd.read_csv("../../Data/ballad_sad20.csv")
sad20
```
