```python
import pandas as pd
import nltk # national language tool kit
from konlpy.corpus import kolaw
from konlpy.tag import Okt
import matplotlib.pyplot as plt
from wordcloud import WordCloud

t = Okt()
```

```python
ballad_indifference20 = pd.read_csv("../../Data/ballad_indifference20.csv")
ballad_indifference20.head()
```

---
### 무관심 wordcloud 생성

```python
indifference20 = ""
```

```python
for i in range(0, 20):
    indifference20 += (" " + ballad_indifference20['가사'][i])
print(indifference20)
```

```python
# 명사 분리
token_indifference = t.nouns(indifference20)
```

```python
# 무관심테마 20곡의 상위 50개 단어 확인
# 이걸 보면서 불용어 선택하기
token_indifference50 = nltk.Text(token_indifference)
token_indifference50.vocab().most_common(50)
```

```python
# 불용어 추가
# 실행해보면서 추가하면서 바꿔야함.
stop_words20 = ['내','말','수','네','날','더','게','것','그','알','젠','속','널','난','일','걸','만큼','해','거','곳','비','좀','덜','뒤','눈','건','손','길']
token_indifference50 = [each_word for each_word in token_indifference if each_word not in stop_words20]
```

```python
# 무관심테마 20곡의 상위 50개 단어 확인
# 불용어 처리후
token_indifference50 = nltk.Text(token_indifference50)
token_indifference50.vocab().most_common(50)
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# 상위 50개의 빈도수 차트 만들기
plt.figure(figsize=(12,6))
token_indifference50.plot(50)
plt.show()
```

```python
# 상위 50개만 wordcloud로 그리기
# 구한데이터로 그리기
data50 = token_indifference50.vocab().most_common(50)

# 딕셔너리로 만들기
tmp_data50 = dict(data50)

wordcloud50 = WordCloud(
    background_color='White',
    relative_scaling=0.2,
    font_path="AppleGothic" # 윈도우 => 'c:/Windows/Fonts/malgun.ttf'
).generate_from_frequencies(tmp_data50)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(wordcloud50, interpolation='bilinear')
plt.axis('off')
plt.show()
```
